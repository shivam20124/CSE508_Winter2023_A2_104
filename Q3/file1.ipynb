```python
import numpy as np
def finding_DCG(df):
    rel_scores = list(df['relevance_scores'])
    DCG_score = 0
    for i in range(50):
        rel_i = rel_scores[i]
        log_val = np.log2(i+2)
        DCG_score = DCG_score + rel_i/log_val
    return DCG_score
```

```python
DCG_ori = finding_DCG(df)
IDCG_val = finding_DCG(sorted_df)
nDCG_val = DCG_ori/ IDCG_val
print(nDCG_val)
```
